c. Which of the following words is the longest: Supercalifragilisticexpialidocious, onorificabilitudinitatibus, or Bababadalgharaghtakamminarronnkonn? 

```python
def longest_word(words):
    longest='' # declare initial variable for longest word as empty
    # iterate through the words
    for num in range(len(words)):
        if len(longest) < len(words[num]):
            longest = words[num]
    print ("longest word is : " + longest)
           
#Ececution
words = ["Supercalifragilisticexpialidocious", "onorificabilitudinitatibus", "Bababadalgharaghtakamminarronnkonn"]     
longest_word(words)
```

d. Which composer comes first in the dictionary: 'Berlioz', 'Borodin', 'Brian', 'Bartok', 'Bellini', 'Buxtehude', 'Bernstein'. Which one comes last?

### Question 3

Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area 
of a triangle with side lengths a, b, and c is  sqtr [s(s - a)(s -b)(s -c)]
, where 
s = (a+b+c)/2

```python
import math
#Execution
def triangleArea(a, b, c):
    # semi-perimeter calculated as thus:
    s = (a + b + c) / 2
    
    # evaluate the value inside the square root
    s_factor = (s * (s - a) * (s - b) * (s - c)) 
    
    # Apply Heron's formula for calculating area
    area = (math.sqrt(s_factor))
    
    return area

# Let's the area using some sample values
print(triangleArea(3, 4, 6))
```

## Question 4
Write a program in python to separate odd and even integers in separate arrays. Go 
to the editor
Test Data : <br>
Input the number of elements to be stored in the array :5<br>
Input 5 elements in the array :<br>
element - 0 : 25<br>
element - 1 : 47<br>
element - 2 : 42<br>
element - 3 : 56<br>
element - 4 : 32<br>
Expected Output:<br>
The Even elements are:<br>
42 56 32<br>
The Odd elements are :<br>
25 47


```python
# Evaluation

# Declare empty lists for the list, even and odd arrays
lst = []
even = []
odd = []

for i in range(5):
    if i == 0:
        num = int(input("Please enter number : 25 \n"))
    if i == 1:
        num = int(input("Please enter number : 47 \n"))
    if i == 2:
        num = int(input("Please enter number : 42 \n"))
    if i == 3:
        num = int(input("Please enter number : 56 \n"))
    if i == 4:
        num = int(input("Please enter number : 32 \n"))
    lst.append(num)
for integers in lst:
    if integers % 2 == 0:
        even.append(integers)
    else:
        odd.append(integers)
            
print("The Even elements are:")
print(even, end = "\n")
print("The Odd elements are:")
print(odd, end = "\n")
```
