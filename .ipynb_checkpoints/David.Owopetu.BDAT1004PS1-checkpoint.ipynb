### Question 3 

Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the are 
of a triangle with side lengths a, b, and c  
s
s(s - a)(s -b)(s c)
, where
s = (a+b+c)/2.
>`>>> triangleArea(2,2,2)` <br>
1.7320508075688772

````python
# Answer to Question 3

import math
#Execution
def triangleArea(a, b, c):
    # semi-perimeter calculated as thus:
    s = (a + b + c) / 2
    
    # evaluate the value inside the square root
    s_factor = (s * (s - a) * (s - b) * (s - c)) 
    
    # Apply Heron's formula for calculating area
    area = (math.sqrt(s_factor))
    
    return area

# Let's the area using some sample values
print(triangleArea(2, 2, 2))
````

### Question 6
You can turn a word into pig-Latin using the following two rules (simplified):<br>
• If the word starts with a consonant, move that letter to the end and append 
'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.<br>
• If the word starts with a vowel, simply append 'way' to the end of the word. 
For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For 
our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).<br>
Write a function pig() that takes a word (i.e., a string) as input and returns its pig-Latin form. Your function should still work if the input word contains upper case 
characters. Your output should always be lower case however. <br>

> ` pig('happy')` <br>
> `'appyhay'`<br>
> `pig('Enter')`<br>
> `'enterway'` 


````python
# Answer to Question 6

# Declare variables
vowels = ["a", "e", "i", "o", "u"]
vowel_word_suffix = "way" # Vovel word suffix - "way"
cons_word_suffix = "ay" # Consonant word suffix "ay"

def pig(word):
    # Handle upper case letter issues
    word = word.lower()
    
    # Capture first letter
    first_letter = word[0]
    
    # Confirm if the word starts with a vowel or consonats
    if first_letter in ['a', 'e', 'i', 'o', 'u']:
        pig_latin = word + vowel_word_suffix
        print(pig_latin)
    else:
        word_length = len(word)
        pig_latin = word[1:word_length] + word[0] + cons_word_suffix
        print(pig_latin)

# Test cases
pig('happy')
pig('Enter')
````

## Question 7 

File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic. 
Write a function bldcount() that reads the file with name name and reports (i.e., 
prints) how many patients there are in each bloodtype. <br>
> `>>>bldcount('bloodtype.txt')`
<br>
There are 10 patients of blood type A. <br>
There is one patient of blood type B. <br>
There are 10 patients of blood type AB. <br>
There are 12 patients of blood type O. <br>
There are no patients of blood type OO.

````python
# Answer to question 7
def bldcount(file_name):
    infile = open(file_name, "r") # read and open file
    content = infile.read()
    lst = content.split() # read file into word
    
    bloodtypes = ['A','B','AB','O','OO']
    
    #iterate to count for each bloodgroup
    for bloodtype in bloodtypes:
        count = lst.count(bloodtype)
        
        # Write statement for one count ("there is ...")
        if count == 1:
            print(f"There is {count} patient of blood type {bloodtype}.")
        
        # Write statement for zero count ("There are no...")
        elif count == 0:
            print(f"There are no patients of blood type {bloodtype}.")
        else:
            print(f"There are {count} patients of blood type {bloodtype}.")

bldcount("bloodtype1.txt")
````

## Question 8

Write a function curconv() that takes as input:<br>
1. a currency represented using a string (e.g., 'JPY' for the Japanese Yen or 
'EUR' for the Euro)
2. an amount and then converts and returns the amount in US dollars.

`>>> curconv('EUR', 100)`<br>
122.96544 <br>
`>>> curconv('JPY', 100)` <br>
1.241401 <br>

The currency rates you will need are stored in file currencies.txt:<br>
AUD 1.0345157 Australian Dollar<br>
CHF 1.0237414 Swiss Franc<br>
CNY 0.1550176 Chinese Yuan<br>
DKK 0.1651442 Danish Krone<br>
EUR 1.2296544 Euro<br>
GBP 1.5550989 British Pound <br>
HKD 0.1270207 Hong Kong Dollar <br>
INR 0.0177643 Indian Rupee<br>
JPY 0.01241401 Japanese Yen<br>
MXN 0.0751848 Mexican Peso<br>
MYR 0.3145411 Malaysian Ringgit <br>
NOK 0.1677063 Norwegian Krone <br>
NZD 0.8003591 New Zealand Dollar <br>
PHP 0.0233234 Philippine Peso <br>
SEK 0.148269 Swedish Krona <br>
SGD 0.788871 Singapore Dollar <br>
THB 0.0313789 Thai Baht <br>

````python
# Answer to question 8

def curconv(cur, amt):
    # Read currency rates from the file
    currency_rates = {}
    with open('currencies.txt', 'r') as file:
        for line in file:
            code, rate, *_ = line.strip().split()
            currency_rates[code] = float(rate)

    # Convert the amount to US dollars
    if cur in currency_rates:
        return amt * currency_rates[cur]
    else:
        return "Currency code not found."

# Test cases
print(curconv('EUR', 100))  # Output: 122.96544
print(curconv('JPY', 100))  # Output: 1.241401
````

### Question 10

Encryption is the process of hiding the meaning of a text by substituting letters in the 
message with other letters, according to some system. If the process is successful, no 
one but the intended recipient can understand the encrypted message. Cryptanalysis
refers to attempts to undo the encryption, even if some details of the encryption are 
unknown (for example, if an encrypted message has been intercepted). The first step 
of cryptanalysis is often to build up a table of letter frequencies in the encrypted text. 
Assume that the string letters is already defined as 
'abcdefghijklmnopqrstuvwxyz'. Write a function called frequencies()
that takes a string as its only parameter, and returns a list of integers, showing the 
number of times each character appears in the text. Your function may ignore any 
characters that are not in letters. <br>

```frequencies('The quick red fox got bored and went home.')``` <br>
[1, 1, 1, 3, 5, 1, 1, 2, 1, 0, 1, 0, 1, 2, 4, 0, 1, 2, 0, 2, 
1, 0, 1, 1, 0, 0] <br>
frequencies('apple')

````python
# Answer to Question 10


def frequencies(text):
    letters = "abcdefghijklmnopqrstuvwxyz" #string of letters
    integers = []
    
    # iterate through the letters
    for char in letters:
        count_char = text.count(char)
        
        # Append to the numbers
        integers.append(count_char)  # Append to the numbers  
    print(integers)               
    

frequencies("The quick red fox got bored and went home.")
frequencies("apple")
````
